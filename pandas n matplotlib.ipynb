```python
import matplotlib.pyplot as plt
import numpy as np  # Importing numpy for mathematical functions
```

```python
x = np.linspace(0, 10, 100)  # Generate 100 points between 0 and 10
y = np.sin(x)  # Calculate the sine of each x value

plt.plot(x, y,linestyle='--', color='r')  # Plot the sine curve
plt.title("Sine Curve")  # Title of the graph
plt.xlabel("X values")  # Label for X-axis
plt.ylabel("Y values (sin(x))")  # Label for Y-axis
plt.grid(True)
plt.legend(['Label for the line'])
plt.show()  # Show the graph
```
